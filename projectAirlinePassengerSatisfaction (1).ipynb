# Outlier Detection

```python
# Define a function to detect outliers using the IQR method
def detect_outliers_IQR(data, column):
    Q1 = np.percentile(data[column], 25)
    Q3 = np.percentile(data[column], 75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return (data[column] < lower_bound) | (data[column] > upper_bound)
```

```python
# Apply outlier detection on numerical columns
numerical_columns = ['age', 'inflight_wifi_service', 'ease_of_online_booking', 'food_and_drink', 'online_boarding',
                     'seat_comfort', 'inflight_entertainment', 'onboard_service', 'leg_room_service',
                     'baggage_handling', 'checkin_service', 'inflight_service', 'cleanliness',
                     'departure_delay_in_minutes', 'arrival_delay_in_minutes']
```

```python
outliers = data[numerical_columns].apply(detect_outliers_IQR, column=0)
print(outliers)
```

```python
# Outlier Handling

# Replace outliers with the 5th and 95th percentile values (a robust method)
for col in numerical_columns:
    lower_percentile = data[col].quantile(0.05)
    upper_percentile = data[col].quantile(0.95)
    data[col] = np.where(data[col] < lower_percentile, lower_percentile, data[col])
    data[col] = np.where(data[col] > upper_percentile, upper_percentile, data[col])
```

```python
# Example of feature engineering: Creating a new feature 'total_delay'
data['total_delay_avg'] = (data['departure_delay_in_minutes'] + data['arrival_delay_in_minutes'])/2
data.drop(['departure_delay_in_minutes', 'arrival_delay_in_minutes'], axis=1, inplace=True)
```

```python
# Calculate the overall facility satisfaction by taking the average of relevant columns
internet_facility_cols = ['inflight_wifi_service', 'online_boarding']  # Add other relevant columns here
data['internet_facility'] = data[internet_facility_cols].mean(axis=1)

# Drop the individual facility satisfaction columns since they are combined into 'overall_facility_satisfaction'
data.drop(internet_facility_cols, axis=1, inplace=True)
```

```python
# Calculate the overall convenience rating
convenience_cols = ['departure_arrival_time_convenient', 'gate_location', 'ease_of_online_booking', 'checkin_service']
data['overall_convenience_rating'] = data[convenience_cols].mean(axis=1)

# Calculate the overall service quality rating
service_quality_cols = ['inflight_service', 'food_and_drink','cleanliness']
data['overall_service_quality_rating'] = data[service_quality_cols].mean(axis=1)

# # Calculate the overall cleanliness rating
# cleanliness_cols = ['cleanliness']
# da['overall_cleanliness_rating'] = df[cleanliness_cols]

# Drop the individual convenience, service quality, and cleanliness rating columns
data.drop(convenience_cols + service_quality_cols, axis=1, inplace=True)
```

```python
# Drop the original 'age' column
data.drop('age', axis=1, inplace=True)
```

```python
# Create distance ranges
def categorize_distance(distance):
    if distance < 500:
        return 'Short Haul'
    elif distance < 1000:
        return 'Medium Haul'
    else:
        return 'Long Haul'

data['Distance_Category'] = data['flight_distance'].apply(categorize_distance)

# Drop the original 'flight_distance' column
data.drop('flight_distance', axis=1, inplace=True)
```
