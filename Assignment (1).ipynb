Display the columns in the dataset.

```python
print(data.columns.tolist())
```

```python
print('SL is',data['SL'].tolist())
print('SW is',data['SW'].tolist())
print('PL is',data['PL'].tolist())
print('PW is',data['PW'].tolist())
print('Classification are',data['Classification'].tolist())
```

Calculate the mean of each column of the dataset

```python
print('Mean of SL is',data["SL"].mean() )
print("Mean of SW is",data["SW"].mean())
print("Mean of PL is",data["PL"].mean())
print("Mean of PW is",data["PW"].mean())
```

Perform meaningful visualizations using the dataset. Bring at least 3
visualizations.


```python
sns.histplot(x='PL', data=data, kde=True, hue='Classification')
plt.title('Histogram')
plt.show()
```

```python
plt.bar(data['Classification'], data['PL'])
plt.title('Bar graph')
plt.xlabel('Classification')
plt.ylabel('PL')
plt.show()
```

```python
sns.lineplot(data=data)
plt.title('Line Graph')
plt.show()
```

```python
sns.lineplot(x='PL', y='PW', data=data)
plt.title('Line Graph')
plt.show()
```
